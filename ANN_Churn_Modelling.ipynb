```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder_1 = LabelEncoder()
X[:, 1] = labelencoder_1.fit_transform(X[:, 1])
labelencoder_2 = LabelEncoder()
X[:, 2] = labelencoder_2.fit_transform(X[:, 2])
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(Xtrain.shape)
print(Xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
Xtrain = sc_X.fit_transform(Xtrain)
Xtest = sc_X.transform(Xtest)
```

# Buildinng the ANN

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# initialising the ANN
classifier = Sequential()
```

```python
# adding the input layer and hidden layer
classifier.add(Dense(6, input_dim = 11, kernel_initializer='glorot_uniform', activation = 'relu'))
```

```python
# adding the second hidden layer
classifier.add(Dense(6, kernel_initializer='glorot_uniform', activation = 'relu'))
```

```python
# adding the 3rd hidden layer
classifier.add(Dense(6, kernel_initializer = 'glorot_uniform', activation = 'relu'))
```

```python
# adding the final output layer
classifier.add(Dense(1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))
```

```python
# compiling the results in our ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# fitting our Neural network model
classifier.fit(Xtrain, ytrain, batch_size = 10, epochs = 100)
```

## accuracy = 86.56%

```python
# predicting the results
y_pred = classifier.predict(Xtest)
```

```python
y_pred = (y_pred > 0.5)
for i, v in zip(ytest, y_pred):
    print(i, v, sep = '---->')
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest, y_pred)
```
